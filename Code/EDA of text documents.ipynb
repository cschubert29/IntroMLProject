```python
#imports

from gensim.corpora import Dictionary
from gensim.corpora import MmCorpus
import operator
import itertools
from collections import defaultdict
import pandas as pd
```

```python
loaded_dct = Dictionary.load('../Data/text-train.dict')
```

```python
word_count = loaded_dct.dfs
sorted_word_count = sorted(word_count.items(), key=operator.itemgetter(1), reverse=True)
```

```python
sorted_word_count
```

```python
#convert list of tuples to dictionary
sorted_dict = dict(sorted_word_count)
```

```python
sorted_dict
```

```python
#token id's
token_id = loaded_dct.token2id
```

```python
token_id
```

```python
def fix_dictionary(dict_):
    return {token_id.get(k, k): v for k, v in dict_.items()}
```

```python
fix_dictionary(sorted_dict)
```
