```python
def flatten_json(y):
    out = {}

    def flatten(x, name=''):
        if type(x) is dict:
            for a in x:
                flatten(x[a], name + a + '_')
        elif type(x) is list:
            i = 0
            for a in x:
                flatten(a, name + str(i) + '_')
                i += 1
        else:
            out[name[:-1]] = x

    flatten(y)
    return out
```

```python
#First remove keys that we won't be using. then apply function as done below. 
#If you don't remove stuff first then you will have way too many extra columns after
```

```python
flat = []
for d in data:
    flat.append(flatten_json(d))
```

```python
from pandas.io.json import json_normalize
json_normalize(flat)
```
