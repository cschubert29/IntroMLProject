```python
#imports
from gensim.corpora import Dictionary
import operator
import matplotlib.pyplot as plt
import numpy as np
import csv
from nltk.corpus import stopwords
```

```python
loaded_dct = Dictionary.load('../Data/text-train.dict')
```

```python
word_count = loaded_dct.dfs
```

```python
word_count_dict = dict(word_count)
```

```python
#token id's
token_id = loaded_dct.token2id
```

```python
#convert to word:frequency format
for k, v in token_id.items():
    coords = word_count_dict[v]
    token_id[k] = coords
```

```python
token_id
```

```python
#sort token_id
sorted_token_id = sorted(token_id.items(), key=lambda kv: kv[1], reverse=True)
```

```python
#Remove stop stords (NLTK)
filter_set = set(stopwords.words('english'))
tuples_filtered = [tup for tup in sorted_token_id if tup[0] not in filter_set]
```

```python
#Remove stop words (legal domain)

legal_words=[]
with open('../Data/legal-stop-words.csv', 'r') as f:
    for row in csv.reader(f):
        legal_words.append(row[0])
        
filter_set1 = set(legal_words)
tuples_filtered1 = [tup for tup in tuples_filtered if tup[0] not in filter_set1]
```

```python

words = list(zip(*tuples_filtered1))[0]
counts = list(zip(*tuples_filtered1))[1]
y_pos = np.arange(len(words)) 
```

```python
#Plot top 50 most frequent words
plt.figure(figsize=(20,10))
plt.barh(y_pos[:50], counts[:50])
plt.yticks(y_pos[:50],words[:50])
plt.gca().invert_yaxis()
plt.title('Top 50 most frequent words used in bill text')
plt.show()
```
